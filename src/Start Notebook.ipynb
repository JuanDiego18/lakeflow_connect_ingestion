```python
  import requests

  DATABRICKS_HOST = dbutils.widgets.get("host") if dbutils.widgets.get("host") else spark.conf.get("spark.databricks.workspaceUrl")
  DATABRICKS_TOKEN = dbutils.secrets.get(scope="jd-scope", key="databricks-token")
  PIPELINE_ID = dbutils.widgets.get("pipeline_id")

  headers = {
      "Authorization": f"Bearer {DATABRICKS_TOKEN}",
      "Content-Type": "application/json"
  }

  response = requests.post(
      f"https://{DATABRICKS_HOST}/api/2.0/pipelines/{PIPELINE_ID}/updates",
      headers=headers,
      json={}
  )

  if response.status_code == 200:
      print(f"Gateway started successfully. Update ID: {response.json().get('update_id')}")
  else:
      raise Exception(f"Failed to start gateway: {response.status_code} - {response.text}")
```
